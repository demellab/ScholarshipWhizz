```python
from langchain_community.embeddings import HuggingFaceEmbeddings
embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
```

```python
#ADDED this code for retriveing data from the CSV file provided by scholarship whizz
CSV_PATH = "scholarship_whizz_db.csv"

def build_vectorstore_from_scholarships(csv_path=CSV_PATH):
    df = pd.read_csv(csv_path)

    # Make sure the column names below EXACTLY match your CSV headers
    required = ["Scholarship Name", "Provider", "Amount", "Closing Date", "Degree(s)", "Status", "Link"]
    missing = [c for c in required if c not in df.columns]
    if missing:
        raise ValueError(f"Missing columns in CSV: {missing}")

    docs = []
    for i, row in df.iterrows():
        # Text that will be indexed for retrieval (edit wording if you like)
        content = (
            f"Scholarship Name: {row['Scholarship Name']}\n"
            f"Provider: {row['Provider']}\n"
            f"Amount: {row['Amount']}\n"
            f"Closing Date: {row['Closing Date']}\n"
            f"Degree(s): {row['Degree(s)']}\n"
            f"Status: {row['Status']}\n"
            f"Link: {row['Link']}"
        )
        docs.append(Document(
            page_content=content,
            metadata={
                "row": int(i),
                "name": row["Scholarship Name"],
                "provider": row["Provider"],
                "closing_date": row["Closing Date"],
                "degree": row["Degree(s)"],
                "link": row["Link"],
            }
        ))

    splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=100)
    chunks = splitter.split_documents(docs)

    vs = Chroma.from_documents(
        chunks,
        embedding=embeddings,
        collection_name="scholarships_csv",
        # persist_directory="./chroma_scholarships"   # ← uncomment to persist to disk
    )
    return vs

# Build once (or rebuild if CSV changed)
vectorstore = build_vectorstore_from_scholarships()
retriever = vectorstore.as_retriever(search_kwargs={"k": 5})
```

```python
#retrieval from the CSV files 
@tool
def retrieve_context(query: str) -> str:
    """
    Retrieve relevant scholarships from the CSV index.
    Returns a readable string the model can use/quote.
    """
    results = retriever.invoke(query)
    lines = []
    for doc in results:
        md = doc.metadata
        lines.append(
            f"- {md.get('name','(no name)')} | Provider: {md.get('provider','-')} | "
            f"Amount: {doc.page_content.split('Amount: ')[1].splitlines()[0] if 'Amount:' in doc.page_content else '-'} | "
            f"Closing: {md.get('closing_date','-')} | Degree(s): {md.get('degree','-')} | Link: {md.get('link','-')}"
        )
    return "\n".join(lines) if lines else "No matching scholarships found."
```

```python
# Bind our retrieval tool and connect it to the model

# List of tools (functions) to use
tools = [retrieve_context]

# Create a ToolNode for LangGraph - run this tool when the agent wants and decides if the tool needs to be run
tool_node = ToolNode(tools)

# Instantiate a Chat model using your internal vLLM backend (OpenAI-compatible)
from langchain_openai import ChatOpenAI

model = ChatOpenAI(
    base_url="http://research-ai.tail69783d.ts.net:8001/v1",  # your internal endpoint
    api_key="EMPTY",  # not needed for local/vLLM servers
    model="openai/gpt-oss-20b",  # adjust if your local model name differs
    temperature=0
).bind_tools(tools)

print("Internal vLLM model connected and tools bound successfully")
```

```python
# Function to decide whether to continue or stop the workflow

def should_continue(state: MessagesState) -> Literal["tools", END]:
    """Determine whether to continue the graph or end the conversation."""
    messages = state["messages"]
    last_message = messages[-1]

    # If the LLM makes a tool call, go to the "tools" node
    if last_message.tool_calls:
        return "tools"

    # Otherwise, finish the workflow
    return END
```

```python
# Function that invokes the model
def call_model(state: MessagesState):
    messages = state["messages"]
    response = model.invoke(messages)
    return {"messages": [response]}  # Returns as a list to add to the state
```

```python
# Define the workflow with LangGraph
workflow = StateGraph(MessagesState)

# Add nodes to the graph
workflow.add_node("agent", call_model)
workflow.add_node("tools", tool_node)

# Connect nodes
workflow.add_edge(START, "agent")
workflow.add_conditional_edges("agent", should_continue)  # Decision after the "agent" node
workflow.add_edge("tools", "agent")  # Cycle between tools and agent
```

```python
checkpointer = MemorySaver()
app = workflow.compile()
```

```python
final_state = app.invoke(
    {
        "messages": [HumanMessage(content="print 5 scholarships in 2025")],
        "configurable": {"thread_id": 42}
    }
)

print(final_state["messages"][-1].content)
```
